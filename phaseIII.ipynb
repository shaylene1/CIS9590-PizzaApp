```python
import numpy as np
import xlrd
from datetime import time
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel(r'C:\Users\Shenm\Desktop\python\PizzaSales.xlsx')
```

```python
totalprice = np.array(data.total_price.astype(int))
pizzasize = np.array(data.pizza_size)
ordertime = np.array([time.hour for time in data.order_time])
orderdate = np.array([date.month for date in data.order_date])
pizzacategory = np.array(data.pizza_category)
pizzaname = np.array(data.pizza_name)
```

```python
num_bins = 100
n, bins, patches = plt.hist(totalprice, num_bins,density = 1,color ='green',alpha = 0.7)

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.title('Distribution of pizza price that customer spend\n\n',fontweight ="bold")

plt.show()
```

```python
plt.bar(pizzasize, range(len(pizzasize)))
plt.title("Is there any preference for pizza size?")
plt.show()
```

```python
unique, count = np.unique(pizzacategory, return_counts=True)
data_count=dict(zip(unique,count))

plt.pie(data_count.values(), labels=data_count.keys(), autopct="%1.1f%%",
 explode=[0,0.1,0,0])
plt.title("Which pizza category is most prederred?")
```

```python
unique, count = np.unique(ordertime, return_counts=True)
data_count=dict(zip(unique,count))

plt.plot(data_count.keys(), data_count.values(), label="ATT-RLSTM")
plt.title("What time of day do people like to order pizza?")
```

```python
unique, count = np.unique(orderdate, return_counts=True)
data_count=dict(zip(unique,count))

plt.plot(data_count.keys(), data_count.values(), label="ATT-RLSTM")
plt.title("What month did people order pizza most?")
```
